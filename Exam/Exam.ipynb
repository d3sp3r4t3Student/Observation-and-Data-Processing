```python
def gauss_2D (x, y, s, c, centerx, centery, sigma):
    g = np.zeros((len(x),len(y)))
    for i in range(len(x)):
        for j in range(len(y)):
            g[i,j] = s*c*np.exp(-0.5*((x[i]-centerx)**2 + (y[j]-centery)**2)/sigma**2)
    return g
```

```python
gauss = lambda x, center, sigma, s, c: s * c * np.exp(-0.5*((x - center)/sigma)**2)

c = 7.15 * 10**16
x = np.arange(0, 10, 0.01)
center = 5
sigma = 1.5
s = 0.07
fig, ax = plt.subplots(figsize=(16,9))
plt.plot(x, gauss(x, center, sigma, s, c))
plt.xlabel('distance in km')
plt.ylabel('NO2 molecules per cm**2')
plt.show()
```
